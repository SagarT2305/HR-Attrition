```python
def anova(df, con, cat):
    import pandas as pd
    from statsmodels.formula.api import ols
    from statsmodels.stats.anova import anova_lm
    p = con + '~'+ cat
    model = ols(p,df).fit()
    anova_results = anova_lm(model)
    q = pd.DataFrame(anova_results)
    a = q['PR(>F)'][cat]
    return round(a,3)
```

```python
for i in cat:
    print("---MonthlyRate ~",i,"-----")
    p = anova(data,"MonthlyRate", i)
    print(p)
```

# Standardization

```python
def preprocessing(df):
    import pandas as pd
    cat = []
    con = []
    for i in df.columns:
        if(df[i].dtypes == "object"):
            cat.append(i)
        else:
            con.append(i)
    X1 = pd.get_dummies(df[cat])
    from sklearn.preprocessing import StandardScaler
    ss = StandardScaler()
    X2 = pd.DataFrame(ss.fit_transform(df[con]),columns=con)
    X3 = X2.join(X1)
    return X3
```

```python
plt.figure(figsize=(12,12))
t=1
for j in ['PercentSalaryHike','YearsSinceLastPromotion', 'JobSatisfaction','MonthlyIncome']:
    from sklearn.cluster import KMeans
    k = list(range(1,15,1))
    WCSS = []
    c1 = data[['YearsAtCompany',j]]

    for i in range(1,15,1):
        km = KMeans(n_clusters = i)
        model = km.fit(c1)
        WCSS.append(model.inertia_)
    
    import matplotlib.pyplot as plt
    plt.subplot(2,2,t)
    plt.scatter(k,WCSS, c='black')
    plt.plot(k,WCSS, c='blue')
    plt.xlabel("K")
    plt.ylabel("WCSS [Within Clusters Sum of Squares]")
    plt.title(j)
    plt.xticks(range(1,15,1))
    plt.grid( axis='both',color='pink', linestyle='-', linewidth=1)
    t = t + 1
 
```
